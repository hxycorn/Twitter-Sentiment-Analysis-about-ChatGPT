```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
query = 'ChatGPT since:2022-12-01 until:2023-01-31 '

limit = 500
```

```python
query = 'ChatGPT until:2023-01-01'
limit = 100000
tweets_1 = []
for tweet in sntwitter.TwitterSearchScraper(query).get_items():
    #print(vars(tweet))
    #break
    if len(tweets_1) == limit:
        break
    else:
        tweets_1.append([tweet.id, 
                       tweet.user.username, 
                       tweet.user.verified, 
                       tweet.user.location,
                       tweet.user.followersCount,
                       tweet.rawContent, 
                       tweet.date,
                       tweet.retweetedTweet,
                       tweet.lang])
df_1 = pd.DataFrame(tweets_1, columns = ['UserId',
                                     'UserName',
                                     'Verified',
                                     'Location',
                                     'Followers',
                                     'Tweet',
                                     'Date', 
                                     'Retweeted',                                
                                     'Language'])
df_1.head()
```

```python
df_1 = pd.DataFrame(tweets_1, columns = ['UserId',
                                     'UserName',
                                     'Verified',
                                     'Location',
                                     'Followers',
                                     'Tweet',
                                     'Timestamp', 
                                     'Retweeted',                                
                                     'Language'])
df_1
```

```python
df_1.to_csv('Twitter_Dec_1.csv',index=False)
```
